## 分词

```python
fin = open('百度停用词表.txt',encoding='utf-8')
stopwords = fin.readlines()
stopwords = [w.strip('\n') for w in stopwords]
fin.close()
stopwordss= ['年', '月','日', '中', '人', '上','都']
stopwords = stopwords + stopwordss
```

```python
# 输出分好词后的的结果，以及词性字典
import jieba
import jieba.analyse
import jieba.posseg as pseg

def jieba_cut(df):
    text=df['full']
    words = pseg.cut(text)
    l_word=[]
    l_flag=[]
    dic = {}
    # rule_words = ['z', 'vn', 'v', 't', 'nz', 'nr', 'ns', 'n', 'l', 'i', 'j', 'an','a']
    rule_words_exclude=['w','x','un']
    for word, flag in words:
        if flag not in rule_words_exclude:
            if word not in stopwords:
                l_word.append(word)
                l_flag.append(flag)
           
    for word in l_flag:
        if word not in dic:
            dic[word] = 1
        else:
            dic[word] = dic[word] + 1

    return l_word,dic

df_news[["cut", "flag"]]= df_news.apply(jieba_cut,axis=1, result_type="expand")
df_news.head()
```

# kmeans 聚类

```python
from sklearn.cluster import MiniBatchKMeans

num_clusters = 10 # need to be selected wisely
kmeans_model = MiniBatchKMeans(n_clusters=num_clusters,
                               init='k-means++',
                               n_init=1,
                               init_size=1000, batch_size=1000, verbose=0, max_iter=1000)
```

```python
kmeans = kmeans_model.fit(vz)
kmeans_clusters = kmeans.predict(vz)
kmeans_distances = kmeans.transform(vz)
```

```python
sorted_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
```
